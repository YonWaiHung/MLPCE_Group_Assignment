**Part A Question 1**

```python
# Importing Pandas to handle the data in the form of rows and columns
import pandas as pd 

# Importing NumPy for numerical and mathematical functions
import numpy as np

# Reading the dataset in the form of .csv file using Pandas
data = pd.read_csv('DataSet.csv')

# Previewing the dataset
data.head(10)
```

```python
# Removing the Unnamed:0 column since it is unnecessary
data.drop('Unnamed: 0', inplace = True, axis = 1)
data.head()
```

```python
# Checking the data types in the dataset
data.dtypes
```

```python
# Statistical summary of the attributes
data.describe()
```

```python
# Replacing '?' values in the dataset
data.replace('?', np.nan, inplace = True)
data.head()
```

```python
# Checking which columns has missing values
data.isnull().sum()
```

```python
# For numerical columns:
# Replacing the NaNs in the dataset with mean value of their column 
meanNormalized = data['normalized-losses'].astype('float').mean(axis = 0)
data['normalized-losses'].replace(np.nan, meanNormalized, inplace = True)

meanBore = data['bore'].astype('float').mean(axis = 0)
data['bore'].replace(np.nan, meanBore, inplace = True)

meanStroke = data['stroke'].astype('float').mean(axis = 0)
data['stroke'].replace(np.nan, meanStroke, inplace = True)

meanHP = data['horsepower'].astype('float').mean(axis = 0)
data['horsepower'].replace(np.nan, meanHP, inplace = True)

meanPeak = data['peak-rpm'].astype('float').mean(axis = 0)
data['peak-rpm'].replace(np.nan, meanPeak, inplace = True)

# Previewing the dataset
data.head()
```

```python
# For categorical columns:

# Checking which category has the highest count
mode = data['num-of-doors'].mode()

# Replacing the NaN values in the column with the mode
data['num-of-doors'].replace(np.nan, 'four', inplace = True)
data.head(10)
```

```python
# For the target variable (Price):
print('Previous data.shape: ', data.shape)
data.dropna(subset=['price'], axis=0, inplace=True)
print('data.shape After: ', data.shape)
```

```python
# Verifying that there is no more missing values in the dataset
data.isnull().sum()
```

```python
# Replacing the data type of these columns to float:
data[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']] =\
data[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']]\
.astype('float')

# Checking data type once again
data.dtypes
```

```python
# Checking the categorical columns in the dataset
catCols = data.select_dtypes(include='O').keys()
print(catCols)
```

```python
# Finding the number of unique labels in the 'fuel-type' column
data['fuel-type'].unique()
```

```python
# Encoding 'fuel-type' to a numerical column
fuel_dummy = pd.get_dummies(data['fuel-type'])
fuel_dummy = fuel_dummy.add_prefix('fuelType-')
fuel_dummy.head()
```

```python
# Concatenating the encoded data frame to the original dataset
data = pd.concat([data, fuel_dummy], axis = 1)
data.head()
```

```python
# Dropping the 'fuel-type' column
data.drop('fuel-type', axis = 1, inplace = True)
data.head()
```

Question 4

```python
# Performing normalization to uniform the variables with different range of values
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numeric_data = data.select_dtypes(include = np.number)
numeric_data = pd.DataFrame(scaler.fit_transform(numeric_data), columns = numeric_data.columns)
numeric_data.head()
```

```python
# Splitting the data into X and y (for regression model)
X = numeric_data.loc[:, ['engine-size']]
y = numeric_data['price']
X.head()
```

```python
# Building the model

# Using a 70-30 train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 4)

# Training and fitting the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# Model predicting on the test set
predictions = model.predict(X_test)

# Calculating model slope and intercept
print("Estimated model slope:" , model.coef_)
print("Estimated model intercept:" , model.intercept_)  
```

```python
# Making predictions with the model
predicted_prices = model.predict(X_test)[0:5]
print(predicted_prices)
```

```python
# Evaluating the model

# Calculating RMSE
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
print("RMSE value: {:.4f}".format(rmse))

# Calculating R2 Score
from sklearn.metrics import r2_score
print ("R2 Score value: {:.4f}".format(r2_score(y_test, predictions)), '\n')

# Plotting Regression Line
mp.scatter(X, y, color = 'blue', label ='Scatter Plot')
mp.plot(X_test, predictions, color = 'black', linewidth = 3, label = 'Regression Line')
mp.title('Relationship between Engine Size and Price')
mp.xlabel('Engine Size')
mp.ylabel('Price')
mp.legend(loc=4)
mp.show()
```
